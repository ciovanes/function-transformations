```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
```

```python
def beautifull():
    plt.grid()
    plt.axhline(0, c="r", linewidth = 1.3)
    plt.axvline(0, c="r", linewidth = 1.3)
```

## Vertical and horizontal displacements

Since $c$ is a constant greater than zero, then the graph:

- $y=f(x)+c$  -> shifts $c$ units up.
- $y=f(x)-c$  -> shifts $c$ units down.
- $y=f(x-c)$  -> shifts $c$ units right.
- $y=f(x+c)$  -> shifts $c$ units left.

```python
def f(x):
    return x**2


n = 1000
x = np.linspace(-10, 10, num=n)
y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y)
beautifull()
plt.title("Original")
```

```python
n = 1000
x = np.linspace(-10, 10, num=n)
c = 10


# y = f(x) + c
plt.subplot(2, 2, 1)
y1 = f(x) + c
plt.plot(x, y1)

beautifull()
plt.title("y = f(x) + c")


# y = f(x) - c
plt.subplot(2, 2, 2)
y2 = f(x) - c
plt.plot(x, y2)

beautifull()
plt.title("y = f(x) - c")


# y = f(x - c)
plt.subplot(2, 2, 3)
y3 = f(x - c)
plt.plot(x, y3)

beautifull()
plt.title("y = f(x - c)")


# y = f(x + c)
plt.subplot(2, 2, 4)
y4 = f(x + c)
plt.plot(x, y4)

beautifull()
plt.title("y = f(x + c)")


plt.tight_layout(pad=1)
```

## Stretching and compressions

Since $c$ is a constant greater than zero, then the graph:

- $y=c \cdot f(x)$ stretches the graph vertically by a factor of $c$.
- $y= \frac{1}{c} \cdot f(x)$ compresses the graph vertically by a factor of $c$.
- $y=f(c \cdot x)$ compresses the graph horizontally by a factor of $c$.
- $y= f(\frac{1}{c} \cdot x )$ stretches the graph horizontally by a factor of $c$.

```python
def f(x):
    return np.sin(x)


x = np.linspace(-15,15, num=n)
y = f((1/1)*x)

fig, ax = plt.subplots()
ax.plot(x,y)
beautifull()
plt.title("Original")
```

```python
n = 1000
x = np.linspace(-15,15, num=n)
c = 3

# y = c * f(x)
plt.subplot(2, 2, 1)
y1 = c * f(x)
plt.plot(x, y1)

beautifull()
plt.title("y = f(x) + c")


# y = 1/c * f(x)
plt.subplot(2, 2, 2)
y2 = 1/c * f(x)
plt.plot(x, y2)

beautifull()
plt.title("y = f(x) - c")


# y = f(c * x)
plt.subplot(2, 2, 3)
y3 = f(c * x)
plt.plot(x, y3)

beautifull()
plt.title("y = f(x - c)")


# y = f(1/c * x)
plt.subplot(2, 2, 4)
y4 = f(1/c * x)
plt.plot(x, y4)

beautifull()
plt.title("y = f(x + c)")


plt.tight_layout(pad=1)
```

## Reflections

- $y=-f(x)$ reflects the graph about the x-axis.
- $y=f(-x)$ reflects the graph about the y-axis.

```python
def f(x):
    return x**3


x = np.linspace(-10,10, num=n)
y = f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
beautifull()
plt.title("Original")
```

```python
# y = -f(x)
plt.subplot(2, 2, 1)
y1 = -f(x)
plt.plot(x, y1)

beautifull()
plt.title("y = f(x) + c")


# y = f(-x)
plt.subplot(2, 2, 2)
y2 = f(-x)
plt.plot(x, y2)

beautifull()
plt.title("y = f(x) - c")
```

## Examples

```python
def f(x):
    return x**2


n = 1000
x = np.linspace(-10, 10, num=n)
c = 30

y = f(x)
y1 = y + c
y2 = y - c

fig, ax = plt.subplots()
ax.plot(x, y, c="deepskyblue")
ax.plot(x, y1, c="skyblue", linestyle = "--")
ax.plot(x, y2, c="steelblue", linestyle = "--")

ax.legend(handles=[mpatches.Patch(color="deepskyblue", label="original"), 
                    mpatches.Patch(color="skyblue", label="shifted up"),
                   mpatches.Patch(color="steelblue", label="shifted down")])

ax.grid()
# beautifull()
```

```python
def f(x):
    return x**2


n = 1000
x = np.linspace(-10, 10, num=n)
c = 5

y = f(x)
y1 = f(x - c)
y2 = f(x + c)

fig, ax = plt.subplots()
ax.plot(x, y, c="deepskyblue")
ax.plot(x, y1, c="skyblue", linestyle="--")
ax.plot(x, y2, c="steelblue", linestyle="--")

ax.legend(handles=[mpatches.Patch(color="deepskyblue", label="original"), 
                    mpatches.Patch(color="skyblue", label="shifted right"),
                   mpatches.Patch(color="steelblue", label="shifted left")])

ax.grid()
# beautifull()
```

```python
def f(x):
    return np.sin(x)


n = 1000
x = np.linspace(-15,15, num=n)
c = 3

y = f((1/1) * x)
y1 = 1/c * y
y2 = c * y

fig, ax = plt.subplots()
ax.plot(x, y, c="deepskyblue")
ax.plot(x, y1, c="skyblue", linestyle = "--")
ax.plot(x, y2, c="steelblue", linestyle = "--")

ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, 
          handles=[mpatches.Patch(color="deepskyblue", label="original"), 
                   mpatches.Patch(color="skyblue", label="compressed vertically"),
                   mpatches.Patch(color="steelblue", label="streched vertically")])

ax.grid()
# beautifull()
```

```python
def f(x):
    return np.sin(x)


n = 1000
x = np.linspace(-15,15, num=n)
c = 3

y = f((1/1)*x)
y1 = f(c * x)
y2 = f(1/c * x)


fig, ax = plt.subplots()
ax.plot(x, y, c="deepskyblue")
ax.plot(x, y1, c="skyblue", linestyle = "--")
ax.plot(x, y2, c="steelblue", linestyle = "--")


ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, 
          handles=[mpatches.Patch(color="deepskyblue", label="original"), 
                   mpatches.Patch(color="skyblue", label="compressed horizontally"),
                   mpatches.Patch(color="steelblue", label="streched horizontally")])

ax.grid()
# beautifull()
```
